```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from collections import Counter

def scrapFun(url):
    try:
        response = requests.get(url, headers={'User-agent': 'your bot 0.1'})
        soup = BeautifulSoup(response.content, 'html.parser')
        authors = soup.find('div', id='gsc_prf_in')
        authorName = authors.text
        print(authorName)
        un_year = soup.find_all('span', class_='gsc_a_h gsc_a_hc gs_ibl')
        all_year = [years.text for years in un_year]
        counting = Counter(all_year)
        counting['Name'] = authorName;
        new_data = pd.DataFrame(counting, index=[1])
        print(new_data)
        try:
            existing_data = pd.read_excel('output.xlsx')
            updated_data = pd.concat([existing_data, new_data], ignore_index=True)
            updated_data.to_excel('output.xlsx', index=False)
            print('Data appended and file updated')
        except FileNotFoundError:
            # If the file does not exist, create a new file
            new_data.to_excel('output.xlsx', index=False)
            print('File created with new data')

    except Exception as e:
        print(f'Error: {e}')
        print('No data found')

    print('Query run')
```

```python
data = pd.read_excel('/content/Google Scholar links.xlsx')
all_list = data['Google Scholar Profile Link']
print(all_list.tolist())
for name in all_list:
  scrapFun(name);
  print('\n')
```
